```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('train.csv')
df.replace("#REF!", pd.NA, inplace=True)

# Basic info
df.info()
df.describe(include='all')  # includes categorical too

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Convert Fare to numeric, coerce errors to NaN
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Now safely fill missing values with mean
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Create a new feature from Cabin (Deck)
df['Deck'] = df['Cabin'].astype(str).str[0]
df['Deck'] = df['Deck'].replace('n', 'Unknown')

df['Survived'] = df['Survived'].fillna(0)  # or some appropriate value
df['Age'] = df['Age'].fillna(df['Age'].median())

# Set the style of the plots
sns.set(style="whitegrid", palette="muted")












```

```python
# Plotting histograms for numeric columns
plt.figure(figsize=(10, 4))
df[['Age', 'Fare']].hist(bins=20, color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# Count plots for categorical features with custom colors
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Survival Count', fontsize=16)
plt.show()
```

```python
# 1. Survival by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='coolwarm')
plt.title('Survival by Class', fontsize=16)
plt.show()
```

```python
# 2. Survival by Sex
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title('Survival by Sex', fontsize=16)
plt.show()
```

```python

# 3. Survival by Embarked Port
plt.figure(figsize=(6, 4))
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set1')
plt.title('Survival by Embarked Port', fontsize=16)
plt.show()
```

```python
# Ensure 'Survived' and 'Age' are numeric
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Age vs Survival
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='Blues')
plt.title('Age Distribution by Survival', fontsize=16)
plt.show()
```

```python
# Fare vs Survival
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df, palette='Purples')
plt.title('Fare Distribution by Survival', fontsize=16)
plt.show()
```

```python
# Pairplot of numerical columns with a customized color palette
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='coolwarm')
plt.show()
```

```python


# 1. Survival by Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df, palette='muted')
plt.title('Survival by Gender', fontsize=16)
plt.show()
```

```python

# 2. Survival by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Spectral')
plt.title('Survival by Class', fontsize=16)
plt.show()
```

```python

# 3. Fare Distribution by Survival
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df, palette='viridis')
plt.title('Fare vs Survival', fontsize=16)
plt.show()
```

```python

# 4. Age Distribution by Survival
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df, palette='YlGnBu')
plt.title('Age vs Survival', fontsize=16)
plt.show()
```

```python


# 5. Correlation Heatmap with improved colors
cols = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, coerce errors to NaN

corr = df[cols].dropna().corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, annot_kws={"size": 12})
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
